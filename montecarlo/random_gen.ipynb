# Random number generation

```python
import sys
sys.path.append('D:/source/repos')
from utilities.std_imports import *
import random as rnd
import scipy.stats as ss
```

### Basic random numbers

```python
rnd.seed(354)

print('\n Real in [0,1] : \n')
for _ in range(5):
    print(rnd.random())
    
print('\n Int in [0,12] : \n')
for _ in range(5):
    print(rnd.randint(0, 12))    
```

### Random from Parametric distributions

```python
print('\n Uniform in [20,30] : \n')
for _ in range(5):
    print(rnd.uniform(20, 30))

print('\n Normal with (m=0, s=1) : \n')
print(np.random.randn(5))

print('\n Exponential with (scale=1) : \n')
exp_gen = ss.expon(scale = 1.0)
for _ in range(5):
    print(exp_gen.rvs())
```

### Random from Probability-based distributions

```python
np.random.choice([0, 1], p=[0.6, 0.4], size=(5, 4))
```

### Random from Empyrical distributions

```python
seed=1
f=0.3
n=50
rand = np.random.RandomState(seed)
data = rand.randn(n)
data[int(f * n):] += 5
data
```

```python
kde = ss.gaussian_kde(data)
idxs = np.linspace(start=data.min(), stop=data.max(), num=200)
values = kde.evaluate(idxs)
plt.figure(figsize=[16,8])
plt.plot(idxs, values);
values
```

```python
kde = ss.gaussian_kde(data)
idxs = np.linspace(start=data.min(), stop=data.max(), num=1000)
values = kde.evaluate(idxs)
for _ in range(5):
    print(rnd.choice(values))
```
